```r
# Change the datatypes of the columns into suitable types.

# Changing the salary to numeric value
ds$Annual.Salary<-as.numeric(gsub("[$,]","",ds$Annual.Salary))

# Changing the bonus to numeric value
ds$Bonus..<-as.numeric(gsub("[%]","",ds$Bonus..))
```

```r
# Changing the dates.

ds$Exit.Date<-as.Date(ds$Exit.Date, format="%m/%d/%Y")
ds$Exit.Date[is.na(ds$Exit.Date)]<-Sys.Date()
print(ds$Exit.Date)
```

```r
ds$Hire.Date<-as.Date(ds$Hire.Date, format="%m/%d/%Y")
print(ds$Hire.Date)
```

```r
# Extract all rows where the department is "Finance."

install.packages("dplyr")
library(dplyr)

# %>% belongs to package dplyr.
# It is a piping operator that is used to pass the output of one function as input to the next.
emp1<-ds %>% filter(Department=="Finance")
print(emp1)
```

```r
# Find how many employees are there in each country.

ds %>% count(Country)
```

```r
# List all employees earning an annual salary greater than $100,000.

ds %>% filter(Annual.Salary>100000)
```

```r
# Find all employees with the job title "Data Scientist."

ds %>% filter(Job.Title=="Data Scientist")
```

```r
# Calculate the average annual salary for each department.

ds %>% group_by(Department) %>% summarise(avg_sal=mean(Annual.Salary, na.rm=TRUE))
# na.rm ignores the null values.
```

```r
# Find the average salary and bonus percentage for each gender.

ds %>% group_by(Gender) %>% summarise(avg_sal=mean(Annual.Salary, na.rm=TRUE), avg_bonus=mean(Bonus.., na.rm=TRUE))
```

```r
# Determine the average age of employees in each business unit.

ds %>% group_by(Business.Unit) %>% summarise (avg_age=mean(Age, na.rm=TRUE))
```

```r
# Create two new columns for first and last names by splitting the full name column.

ds %>% separate(Full.Name, into=c("First Name", "Last Name"), sep=" ")
```

```r
# Create a new column that calculates the number of years an employee has worked (from hire date to exit date or today if still employed).

# Time difference in days
ds$Years.Worked<-difftime(ds$Exit.Date, ds$Hire.Date)

# Time difference in years
(ds$Years.Worked)/365
```

```r
# Identify the top 5 employees with the longest service duration.

ds %>% arrange(desc(Years.Worked)) %>% slice_head(n=5)
```

```r
# Find employees with a bonus percentage greater than 10% and an annual salary below $80,000.

ds %>% filter(Bonus..>10 & Annual.Salary>80000)
```

```r
# Count the number of unique ethnicities represented in each department.

ds %>% group_by(Department) %>% summarise (unique_ethnicities=n_distinct(Ethnicity))
```

```r
# Find the top 3 departments where the total compensation (annual salary + bonus) is the highest, and display their total compensation.

ds %>%
  mutate(total_compensation = Annual.Salary + Bonus..) %>%  # Create total_compensation here
  group_by(Department) %>%
  summarise(total_compensation = sum(total_compensation, na.rm = TRUE)) %>%
  arrange(desc(total_compensation)) %>%
  slice_head(n = 3)
```

```r
# Calculate the retention rate for each combination of gender and ethnicity. Retention rate is the percentage of employees still working (i.e., exit date is missing).

ds %>%
  mutate(is_retained = is.na(Years.Worked)) %>%
  group_by(Gender, Ethnicity) %>%
  summarise(retention_rate = mean(is_retained) * 100) %>%
  arrange(desc(retention_rate))
```

```r
# For each department, identify the most common job title.

 ds %>%
  group_by(Department, Job.Title) %>%
  summarise(count = n()) %>%
  slice_max(count, n = 1, with_ties = FALSE) %>%
  select(Department, Job.Title)
```

```r
# Calculate the average time (in years) employees worked before exiting for each department.

ds %>%
  filter(!is.na(Exit.Date)) %>%
  mutate(Hire.Date, Exit.Date) %>%
  mutate(time_to_exit = as.numeric(difftime(Exit.Date,Hire.Date, units = "days")) / 365) %>%
  group_by(Department) %>%
  summarise(avg_time_to_exit = mean(time_to_exit, na.rm = TRUE))
```

```r
# Compare the average salary for each job title between genders.

ds %>%
  group_by(Job.Title, Gender) %>%
  summarise(avg_salary = mean(Annual.Salary, na.rm = TRUE)) %>%
  spread(key = Gender, value = avg_salary) %>%
  mutate(pay_gap = abs(`Male` - `Female`)) %>%
  filter(pay_gap > 10000)
```
